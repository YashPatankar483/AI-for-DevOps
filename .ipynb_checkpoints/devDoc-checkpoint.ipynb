```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import tensorflow as tf

from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('devops_errors_dataset.csv')
```

```python
data.isna().sum().sum()
```

```python
X = data.drop('Solution', axis = 1)
y = data['Solution']
```

```python
display(X.head())
display(y.head())
```

```python
columns = X.columns
columns
```

```python
def integrit_all(row):
  error_msg = f"{row['Error Code']} {row['Error Message']} {row['Process']} {row['Environment Details']} {row['Frequency']}"
  return error_msg
```

```python
X['combined_features'] = X.apply(integrit_all, axis = 1)
```

```python
X.head()
```

```python
X_new = X['combined_features']
X_new
```

```python
errors_list = ["solve: " + error for error in X_new]
error_solutions = y.to_list()
```
